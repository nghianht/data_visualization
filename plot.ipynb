**Pie Chart**

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == 'Right'].count()[0]

plt.title('Foot prefernce of players in FIFA')
plt.pie([left,right], labels=['left','right'], autopct = '%.2f')
plt.savefig('/content/drive/My Drive/matplotlib/Foot_preference.png', dpi = 200)
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight ]
```

```python
light = fifa.loc[fifa["Weight"] < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
light_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]
weights_for_chart = [light, light_medium, medium, light_heavy, heavy]
weights_label = ['light', 'light_medium', 'medium', 'light_heavy', 'heavy']
fig = plt.gcf()
explode = [.1,.1,.1,.1,.1]
fig.set_size_inches(10,10) # or (4,4) or (5,5) or whatever
plt.pie(weights_for_chart, labels= weights_label, autopct='%.2f %%', explode= explode)
plt.savefig('/content/drive/My Drive/matplotlib/Weight.png', dpi = 200)
plt.title('Weight distribution of Player in FIFA (in lbs)')
plt.show()
```

**Work with club**

```python
plt.figure(figsize=(6,10))

liverpool = fifa.loc[fifa['Club'] == "Liverpool"]['Overall']
mandan = fifa.loc[fifa['Club'] == "Manchester United"]['Overall']
chelshit = fifa.loc[fifa['Club'] == "Chelsea"]['Overall']

labels = ["Liverpool", "MNGU", 'Chelshit']
plt.title("Professional Clubs Comparison")
plt.ylabel('Fifa Overall')

plt.boxplot([liverpool, mandan, chelshit], labels= labels)
plt.savefig('/content/drive/My Drive/matplotlib/Weight.png', dpi = 200)
plt.show()
```
